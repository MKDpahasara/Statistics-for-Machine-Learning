#  Log Normal Distribution

The Log Normal distribution is a continuous probability distribution of a random variable whose logarithm is normally distributed.

```python
import numpy as np
import matplotlib.pyplot as plt

# Normal distribution parameters
mean = 0  # Mean of the normal distribution
std_dev = 1  # Standard deviation of the normal distribution

# Log Normal distribution parameters
mu = 0  # Mean of the underlying normal distribution (logarithm)
sigma = 1  # Standard deviation of the underlying normal distribution (logarithm)
sample_size = 1000  # Number of samples to generate

# Generate random samples from Normal and Log Normal distributions
normal_data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)
log_normal_data = np.random.lognormal(mean=mu, sigma=sigma, size=sample_size)

# Plotting both distributions for comparison
plt.figure(figsize=(8, 6))

# Plotting Normal distribution
plt.hist(normal_data, bins=30, density=True, alpha=0.5, color='blue', label='Normal Distribution')

# Plotting Log Normal distribution
plt.hist(log_normal_data, bins=30, density=True, alpha=0.5, color='orange', label='Log Normal Distribution')

plt.title('Comparison: Normal vs Log Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```
