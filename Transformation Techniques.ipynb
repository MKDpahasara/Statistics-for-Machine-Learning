# All Transformation Techniques 

Transformation techniques in statistics and data analysis are used to modify data or distributions to meet certain assumptions or improve model performance. Here are some common transformation techniques


**Log Transformation**: Converts data using the logarithm function. Useful for handling skewed data, stabilizing variance, and making relationships more linear.

**Square Root Transformation**: Takes the square root of data values to reduce skewness and stabilize variance.

**Exponential Transformation**: Raises data values to an exponent. Useful for handling growth rates or exponential relationships.

**Box-Cox Transformation**: A family of power transformations that includes the log transformation (when the power is 0) and addresses varying powers to find the best transformation for normalizing data.

**Yeo-Johnson Transformation**: An extension of the Box-Cox transformation that works with both positive and negative values.

**Rank Transformation**: Converts data into ranks or percentiles to reduce the impact of outliers and make the data distribution more uniform.

**Z-Score Transformation (Standardization)**: Scales data to have a mean of zero and a standard deviation of one. Useful for comparing different datasets on the same scale.

**Min-Max Scaling (Normalization)**: Scales data to a fixed range, typically between 0 and 1, maintaining the relationship between values.

**Power Transformation**: Applies a power function to the data to adjust the shape of the distribution.

**Centering**: Shifting the mean of the data to zero by subtracting the mean value from each data point.

**Decomposition**: Breaking down a time series into its components (trend, seasonality, and residuals) using techniques like Seasonal Decomposition of Time Series (STL) or Singular Value Decomposition (SVD).

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
data = np.random.normal(loc=100, scale=20, size=1000)

transformations = {
    'Original Data': data,
    'Log Transformation': np.log(data - min(data) + 1),  # Adding 1 to handle zero/negative values
    'Square Root Transformation': np.sqrt(data - min(data)),
    'Exponential Transformation': np.exp((data - min(data)) / 100),
    'Box-Cox Transformation': np.power(data, 0.5),  # For demonstration purposes; use scipy.stats.boxcox for optimal lambda
    'Rank Transformation': np.argsort(np.argsort(data)) / len(data),
    'Z-Score Transformation': (data - np.mean(data)) / np.std(data),
    'Min-Max Scaling': (data - min(data)) / (max(data) - min(data)),
    'Power Transformation': np.power(data, 2),
}


plt.figure(figsize=(12, 10))
for i, (transformation_name, transformed_data) in enumerate(transformations.items(), 1):
    plt.subplot(3, 3, i)
    plt.hist(transformed_data, bins=30, alpha=0.7)
    plt.title(transformation_name)

plt.tight_layout()
plt.show()
```
