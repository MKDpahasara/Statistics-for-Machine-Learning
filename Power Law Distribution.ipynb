# Power Law Distribution

**Power Law distribution** is a type of continuous probability distribution that represents a relationship between two quantities, where a change in one quantity results in a proportional change in another quantity, following a power-law relationship. This distribution is often used to describe various natural and social phenomena, such as the distribution of city sizes, income distribution, word frequency in languages, and the distribution of earthquakes' magnitudes.

```python
import numpy as np
import matplotlib.pyplot as plt

# Power Law distribution parameters
alpha = 2.5  # Exponent parameter
sample_size = 1000  # Number of samples to generate

# Generate random samples from a Power Law distribution
power_law_data = np.random.pareto(alpha, size=sample_size)


plt.figure(figsize=(8, 6))
plt.hist(power_law_data, bins=30, density=True, alpha=0.7, color='skyblue')

plt.title('Power Law Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
