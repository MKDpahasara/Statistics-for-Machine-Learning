# Probability Distribution Function

***Probability Distribution Function*** can refer to both continuous and discrete distributions, encompassing Probability Density Function (PDF) for continuous distributions and Probability Mass Function (PMF) for discrete distributions.



***Probability Density Function (PDF)-*** Describes the probability distribution of a continuous random variable. It represents the probability of a random variable taking on a particular value or falling within a range of values. The integral over a range gives the probability of the random variable falling within that range.


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
mean = 0
std_dev = 1
x = np.linspace(-5, 5, 1000)
pdf = norm.pdf(x, mean, std_dev)  # Calculate the PDF values
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label='Normal Distribution (μ=0, σ=1)')
plt.title('Probability Density Function (PDF) of a Normal Distribution')
plt.xlabel('X-axis')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

***Probability Mass Function (PMF)-*** Describes the probability distribution of a discrete random variable. It gives the probability that a discrete random variable is exactly equal to a particular value.

```python
num_rolls = 1000
die_rolls = np.random.randint(1, 7, size=num_rolls)
```

```python
unique, counts = np.unique(die_rolls, return_counts=True)
probability = counts / num_rolls
```

```python
plt.figure(figsize=(8, 5))
plt.bar(unique, probability, color='skyblue', edgecolor='black')
plt.title('Probability Mass Function (PMF) of Rolling a Fair Six-sided Die')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.xticks(np.arange(1, 7)
plt.ylim(0, 0.25)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
