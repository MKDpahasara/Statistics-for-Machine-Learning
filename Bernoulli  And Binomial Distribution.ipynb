# Bernoulli Distribution And Binomial

**Bernoulli** deals with the outcome of the single trial of the event, whereas **Binomial** deals with the outcome of the multiple trials of the single event. Bernoulli is used when the outcome of an event is required for only one time, whereas the Binomial is used when the outcome of an event is required multiple times.


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Bernoulli Distribution

```python
from scipy.stats import bernoulli


p = 0.3  # Probability of success

# Create a Bernoulli distribution
bernoulli_dist = bernoulli(p)

# Generate random samples from the Bernoulli distribution
samples = bernoulli_dist.rvs(size=1000)  # Generating 1000 samples


plt.hist(samples, bins=[-0.5, 0.5, 1.5], align='mid', rwidth=0.5, color='skyblue', alpha=0.7)
plt.title('Bernoulli Distribution')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.xticks([0, 1], ['Failure (0)', 'Success (1)'])
plt.grid(axis='y')
plt.show()
```

## Binomial Distribution 

```python
from scipy.stats import binom


n = 10  # Number of trials
p = 0.5  # Probability of success in each trial

binomial_dist = binom(n, p)

samples = binomial_dist.rvs(size=1000) 


plt.hist(samples, bins=np.arange(0, n + 2) - 0.5, align='mid', rwidth=0.8, color='lightgreen', alpha=0.7)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(np.arange(0, n + 1))
plt.grid(axis='y')
plt.show()
```
