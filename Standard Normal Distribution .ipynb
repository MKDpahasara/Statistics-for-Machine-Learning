# Standard Normal Distribution 

The Standard Normal Distribution, also known as the Z-distribution or Gaussian distribution with a mean of 0 and standard deviation of 1, is a special case of the normal distribution. It is characterized by its bell-shaped curve and is commonly used in statistics and probability theory due to its various properties.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
x = np.linspace(-4, 4, 1000)  # Range of x-values
pdf = norm.pdf(x, loc=0, scale=1)  # Probability Density Function of the standard normal distribution
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label='Standard Normal Distribution')
plt.title('Probability Density Function (PDF) of Standard Normal Distribution')
plt.xlabel('X-axis')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
