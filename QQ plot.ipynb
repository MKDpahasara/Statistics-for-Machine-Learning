# QQ plot

**QQ plot**  is a scatterplot created by plotting two sets of quantiles against one another. If both sets of quantiles came from the same distribution, we should see the points forming a line that's roughly straight. Here's an example of a normal QQ plot when both sets of quantiles truly come from normal distributionsns
a QQ plot provides a quick visual way to compare the distribution of a dataset with a theoretical distribution, allowing you to assess whether the dataset fits the assumed distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
np.random.seed(0)  # # Generate sample data
data = np.random.normal(loc=50, scale=10, size=1000)  # Example normal distribution data
```

```python
plt.figure(figsize=(6, 6))
stats.probplot(data, dist="norm", plot=plt)
plt.title('QQ Plot - Normal Distribution')
plt.xlabel('Theoretical quantiles')
plt.ylabel('Sample quantiles')
plt.show()
```
